```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot y = sin(x)
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y_sin = np.sin(x)
ax1.plot(x, y_sin, color='blue')
ax1.set_title("y = sin(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_ylim(-1.5, 1.5)
ax1.set_xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi])
ax1.set_xticklabels(['-2π', '-π', '0', 'π', '2π'])
ax1.axhline(0, color='black',linewidth=0.5)
ax1.axvline(0, color='black',linewidth=0.5)

# Plot y = arcsin(x)
x_arcsin = np.linspace(-1, 1, 500)
y_arcsin = np.arcsin(x_arcsin)
ax2.plot(x_arcsin, y_arcsin, color='red')
ax2.set_title("y = arcsin(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_ylim(-np.pi / 2, np.pi / 2)
ax2.set_yticks([-np.pi / 2, 0, np.pi / 2])
ax2.set_yticklabels(['-π/2', '0', 'π/2'])
ax2.axhline(0, color='black',linewidth=0.5)
ax2.axvline(0, color='black',linewidth=0.5)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up x values for each inverse function based on their domains
x_arcsin_arccos = np.linspace(-1, 1, 500)  # Domain [-1, 1] for arcsin and arccos
x_arctan = np.linspace(-10, 10, 500)       # Wider domain for arctan

# Compute y values for each inverse trig function
y_arcsin = np.arcsin(x_arcsin_arccos)
y_arccos = np.arccos(x_arcsin_arccos)
y_arctan = np.arctan(x_arctan)

# Set up the figure and subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Plot y = arcsin(x)
ax1.plot(x_arcsin_arccos, y_arcsin, color='blue')
ax1.set_title("y = arcsin(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_ylim(-np.pi/2, np.pi/2)
ax1.set_yticks([-np.pi/2, 0, np.pi/2])
ax1.set_yticklabels(['-π/2', '0', 'π/2'])
ax1.axhline(0, color='black',linewidth=0.5)
ax1.axvline(0, color='black',linewidth=0.5)

# Plot y = arccos(x)
ax2.plot(x_arcsin_arccos, y_arccos, color='green')
ax2.set_title("y = arccos(x)")
ax2.set_xlabel("x")
ax2.set_ylim(0, np.pi)
ax2.set_yticks([0, np.pi/2, np.pi])
ax2.set_yticklabels(['0', 'π/2', 'π'])
ax2.axhline(0, color='black',linewidth=0.5)
ax2.axvline(0, color='black',linewidth=0.5)

# Plot y = arctan(x)
ax3.plot(x_arctan, y_arctan, color='red')
ax3.set_title("y = arctan(x)")
ax3.set_xlabel("x")
ax3.set_ylim(-np.pi/2, np.pi/2)
ax3.set_yticks([-np.pi/2, 0, np.pi/2])
ax3.set_yticklabels(['-π/2', '0', 'π/2'])
ax3.axhline(0, color='black',linewidth=0.5)
ax3.axvline(0, color='black',linewidth=0.5)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
